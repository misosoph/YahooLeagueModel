```python
passing = data.loc[data.PlayType.isin(['PASS']) & (data.IsReciever==0) & (data.IsPenalty==0),['GameDate','Quarter','Minute','Second','OffenseTeam', 'Player', 'Yards', 'IsTouchdown', 'IsInterception', 'IsIncomplete', 'IsFumble']]
recieving = data.loc[data.PlayType.isin(['PASS']) & (data.IsReciever==1) & (data.IsPenalty==0) & (data.IsInterception==0),['GameDate','Quarter','Minute','Second', 'OffenseTeam', 'Player', 'Yards', 'IsTouchdown', 'IsFumble']]
rushing = data.loc[data.PlayType.isin(['RUSH', 'SCRAMBLE', 'SACK']) & (data.IsReciever==0) & (data.IsPenalty==0),['GameDate','Quarter','Minute','Second', 'OffenseTeam', 'Player', 'Yards', 'IsTouchdown', 'IsFumble']]
```

```python
play_key = ['GameDate','Quarter','Minute','Second','OffenseTeam','Player']
plays = pd.merge(pd.merge(passing[play_key+['Yards', 'IsTouchdown', 'IsFumble', 'IsInterception', 'IsIncomplete']].rename(columns={'Yards':'Passing', 'IsTouchdown':'PassingTD', 'IsFumble': 'PassingFumble'}).assign(Passes=1), 
                 recieving[play_key+['Yards', 'IsTouchdown', 'IsFumble']].rename(columns={'Yards':'Recieving', 'IsTouchdown':'RecievingTD', 'IsFumble': 'RecievingFumble'}).assign(Receptions=1), on=play_key, how='outer'),
                 rushing[play_key+['Yards', 'IsTouchdown', 'IsFumble']].rename(columns={'Yards':'Rushing', 'IsTouchdown':'RushingTD', 'IsFumble': 'RushingFumble'}), on=play_key, how='outer').fillna(0)
```

```python
gameGroup = plays.groupby([ 'OffenseTeam', 'Player', 'GameDate'])
gameTotals = gameGroup.sum().drop(columns=['Quarter','Minute','Second'])
gameTotals['Fumbles'] = gameTotals.PassingFumble+gameTotals.RecievingFumble+gameTotals.RushingFumble
gameTotals.drop(columns=['PassingFumble', 'RecievingFumble', 'RushingFumble'], inplace=True)
gameTotals['Plays'] = gameGroup.size()
```

```python
def getGameScore(game):
    score = 0
    score += game.Passing/25
    if game.Passing >= 400:
        score += 3
    elif game.Passing >= 300:
        score += 1
    score += game.PassingTD * 4
    score += game.IsInterception * -2
    
    score += game.Rushing/10
    if game.Rushing >= 200:
        score += 3
    elif game.Rushing >= 100:
        score += 1
    score += game.RushingTD * 6    
    
    score += game.Receptions * 1
    score += game.Recieving/10
    score += game.RecievingTD * 6    
    
    return score
```

```python
gameTotals = gameTotals.assign(Score=lambda x: gameTotals.apply(getGameScore, axis=1))
```

```python
playerStats = gameTotals.groupby(['OffenseTeam','Player']).mean().sort_values('Score', ascending=False)
```
