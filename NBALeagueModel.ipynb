```python
def get_winrates(team_stats):
    import pandas as pd
    import numpy as np
    import scipy.stats
    scored = ['FG%','FT%','3PTM','PTS','REB','AST','ST','BLK','TO']
    team_means = team_stats.groupby('OWNER').mean().loc[:,scored]
    team_means['FG%']=team_stats.groupby('OWNER').apply(lambda x:(x['FG%']*x.FGA).sum()/x.FGA.sum())
    team_means['FT%']=team_stats.groupby('OWNER').apply(lambda x:(x['FT%']*x.FTA).sum()/x.FTA.sum())
    team_norm = team_means.apply(lambda x:(x-x.mean())/x.std()).assign(total=lambda x:x.sum(axis=1))
    expected_scores = pd.MultiIndex.from_product([team_norm.index, team_norm.index]).rename(['A','B']).to_frame().transpose().apply(lambda x:np.sum(scipy.stats.norm.cdf(team_norm.loc[x[0],:'TO']-team_norm.loc[x[1],:'TO'])-.5)*2).unstack().transpose()
    expected_win_rate = (expected_scores).apply(scipy.stats.norm.cdf).sum() / len(expected_scores.index)
    return expected_win_rate.sort_values(ascending=False)
```

```python
import multiprocess as mp
from itertools import repeat

def get_replacement_winrates(player, replacements, playerStats, team_list, get_winrates, get_team_stats):
    scored = ['FG%','FT%','3PTM','PTS','REB','AST','ST','BLK','TO']
    
    winrates = [(get_winrates(get_team_stats(team_list.replace(player, replacement), playerStats))["JL's Fine Team"], replacement+" for "+player) for 
             replacement in replacements]
    return winrates

p = mp.Pool(8)

team_list_inj=[]
for owner, players in teams.items():
    team_list_inj.extend([{'OWNER':owner,'PLAYER':player} for player in players if not player in playerFrame[playerFrame.status=='INJ'].name.to_list()])
team_list_inj = pd.DataFrame(team_list_inj)
team_list_inj

alternatives = [item for sublist in [team_list.loc[team_list.OWNER==owner][7:].PLAYER for owner in team_list_inj.OWNER.unique() if owner!="JL's Fine Team"] for item in sublist]
winrates = p.starmap(get_replacement_winrates, zip(team_list.loc[team_list.OWNER=="JL's Fine Team"][4:8].PLAYER, repeat(alternatives), repeat(playerStats), repeat(team_list), repeat(get_winrates), repeat(get_team_stats)))
p.close()
p.join()
```
