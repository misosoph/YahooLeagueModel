```python
def get_team_stats(team_list, playerStats):
    return team_list.set_index('PLAYER').join(playerStats.set_index('PLAYER')).reset_index()
```

```python
def get_winrates(team_stats):
    team_means = team_stats.groupby('OWNER').mean().loc[:,scored]
    team_means['FG%']=team_stats.groupby('OWNER').apply(lambda x:(x['FG%']*x.FGA).sum()/x.FGA.sum())
    team_means['FT%']=team_stats.groupby('OWNER').apply(lambda x:(x['FT%']*x.FTA).sum()/x.FTA.sum())
    team_norm = team_means.apply(lambda x:(x-x.mean())/x.std()).assign(total=lambda x:x.sum(axis=1))
    expected_scores = pd.MultiIndex.from_product([team_norm.index, team_norm.index]).to_frame().transpose().apply(lambda x:np.sum(scipy.stats.norm.cdf(team_norm.loc[x[0],:'TO']-team_norm.loc[x[1],:'TO'])-.5)*2).unstack().transpose()
    expected_win_rate = (expected_scores).apply(scipy.stats.norm.cdf).sum() / len(expected_scores.index)
    return expected_win_rate.sort_values(ascending=False)
```
